```python
import pytrends
from pytrends.request import TrendReq
palabra_clave = input("Palabra a buscar? ")
pytrends = TrendReq(hl="en-US", tz=360)
pytrends.build_payload(kw_list=[palabra_clave])
df = pytrends.interest_over_time()
df[palabra_clave].plot(figsize=(20,7),color='purple',linewidth=7)
```
